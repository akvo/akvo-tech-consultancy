```python
def set_id(x):
    code = x['q'].split(' ')[0].replace('.','|').upper()
    dependency = None
    havedependency = False
    number = 0
    for c in code.split('|'):
        if not havedependency and c in string.ascii_uppercase:
            havedependency = True
    if havedependency:
        number = 0
        if 'I' in code:
            
        dependency = code.split('|')
        dependency = dependency[0]
    code = code.replace('|','')
    x.update({'i': code,'d':dependency})
    for i,o in enumerate(x['a']):
        o.update({'i':'{}#{}'.format(x['i'],i)})
    return x
```

```python
getcode = lambda x: [set_id(item) for item in x]
```

```python
data = r.get('http://unep.localhost/json/unep-forms-with-variables.json').json()
```

```python
data = getcode(data)
```

```python
pd.DataFrame(data)
```

```python
factor = []
for d in data:
    answers = []
    for a in d['a']:
        answers.append('['+ a['i'] +'] '+ a['s'])
    factor.append({
        'Id':d['i'],
        'Dependency':None,
        'DependencyAnswer':None,
        'Question':d['q'],
        'Dropdown':d['s'],
        'Type':d['t'],
        'Answers':'||'.join(answers),
        'Options':d['a']
    })
```

```python
factor = pd.DataFrame(factor)
```

```python
factor = factor[['Id','Dependency','DependencyAnswer','Question','Dropdown','Type','Options','Answers']]
```
