```python
import pandas as pd
```

```python
data = pd.read_csv('pari_hiu_lv3.csv', names=['lv1-code','lv1-name','lv2-code','lv2-name','lv3-code','lv3-name'])
```

```python
data['lv3-code'] = data['lv2-code']
```

```python
data['lv2-name'] = data['lv2-name'].apply(lambda x: x.split('(')[0].title())
```

```python
data['lv3-code'] = data['lv1-code'] + '-' + data['lv3-code']
```

```python
data = data[['lv1-code','lv1-name','lv2-code','lv2-name','lv3-code','lv3-name']]
```
